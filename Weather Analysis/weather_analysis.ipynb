# Spearman CORRELACTION WEATHER AND BLOOD COLLECTIONS

```python
import pandas as pd
import numpy as np

blood_df = pd.read_csv('concelho_metropolitana_lisboa.csv')
weather_df = pd.read_csv('weather_lisbon.csv')
weather_df.rename(columns={'date_time': 'data_saida'}, inplace=True)

blood_df['data_saida'] = pd.to_datetime(blood_df['data_saida'])
weather_df['data_saida'] = pd.to_datetime(weather_df['data_saida'])
group_by_Day = blood_df.groupby(pd.Grouper(key='data_saida', freq="D")).size().reset_index(name='count')

merged_df = pd.merge(group_by_Day, weather_df, on='data_saida', how='left')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

try:
    merged_df = merged_df.drop(['data_saida', 'moonrise', 'moonset', 'sunrise', 'sunset', 'uvIndex.1', 'tempC'], inplace=False, axis=1)
except:
    print('Columns already dropped')

correlated_matrix_ls = merged_df.corr(method='spearman')

# Create heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(correlated_matrix_ls, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
plt.show()
```

```python
# Drop unnecessary columns
try:
    merged_df = merged_df.drop(['data_saida', 'moonrise', 'moonset', 'sunrise', 'sunset', 'uvIndex.1', 'tempC'], inplace=False, axis=1)
except:
    print('Columns already dropped')

# Initialize a dictionary to store lagged cross-correlation coefficients
lagged_correlations = {}

# Calculate cross-correlation for lag values from 1 to 7
for lag in range(0, 3):
    lagged_df = merged_df.copy()
    lagged_df['count'] = lagged_df['count'].shift(+lag)  # Shift blood collections data by lag
    
    # Calculate Spearman correlation matrix
    correlated_matrix_ls = lagged_df.corr(method='spearman')
    
    # Store the correlation coefficient between 'count' and other variables for this lag
    lagged_correlations[lag] = correlated_matrix_ls['count']

# Convert the dictionary to a DataFrame
lagged_correlations_df = pd.DataFrame(lagged_correlations)

# Create heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(lagged_correlations_df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
plt.xlabel('Lag (days)')
plt.ylabel('Variables')
plt.show()
```

```python
import pandas as pd
import numpy as np

blood_df = pd.read_csv('concelho_metropolitana_lisboa.csv')
weather_df = pd.read_csv('weather_lisbon.csv')
weather_df.rename(columns={'date_time': 'data_saida'}, inplace=True)

blood_df['data_saida'] = pd.to_datetime(blood_df['data_saida'])
weather_df['data_saida'] = pd.to_datetime(weather_df['data_saida'])
group_by_Month = blood_df.groupby(pd.Grouper(key='data_saida', freq="M")).size().reset_index(name='count')

pivot = pd.pivot_table(weather_df, 
                       index=pd.Grouper(key='data_saida', freq="M"),
                       values=['mintempC', 'maxtempC', 'precipMM', 'totalSnow_cm', 'sunHour', 'uvIndex', 'moon_illumination', 'DewPointC', 'FeelsLikeC', 'HeatIndexC', 'WindChillC', 'WindGustKmph', 'cloudcover', 'humidity', 'pressure', 'visibility', 'winddirDegree', 'windspeedKmph'], 
                       aggfunc={'mintempC': 'mean', 'maxtempC': 'mean', 'precipMM': 'sum', 'totalSnow_cm': 'sum', 'sunHour': 'mean', 'uvIndex': 'mean', 'moon_illumination': 'mean', 'DewPointC': 'mean', 'FeelsLikeC': 'mean', 'HeatIndexC': 'mean', 'WindChillC': 'mean', 'WindGustKmph': 'mean', 'cloudcover': 'mean', 'humidity': 'mean', 'pressure': 'mean', 'visibility': 'mean', 'winddirDegree': 'mean', 'windspeedKmph': 'mean'})


merged_df = pd.merge(group_by_Month, pivot, on='data_saida', how='left')
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot blood collections as bar plot
ax1.plot(merged_df['data_saida'], merged_df['count'], label='Blood Collections' ,color='lightcoral', alpha=0.5)

# Set labels for the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Blood Collections')

# Create a secondary y-axis for temperature
ax2 = ax1.twinx()

# Plot min and max temperatures as lines
ax2.plot(merged_df['data_saida'], merged_df['mintempC'], label='Min Temp', color='blue')
ax2.plot(merged_df['data_saida'], merged_df['maxtempC'], label='Max Temp', color='orange')

# Set labels for the right y-axis
ax2.set_ylabel('Temperature (°C)')

# Formatting the x-axis dates
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_major_locator(mdates.YearLocator())

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
